````python
def load_json(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
        return data
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return None
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
        return None

file_path = 'code_alpaca_2k.json' 
data = load_json(file_path)

if data:
    for entry in data:
        print(f"Instruction: {entry['instruction']}")
        print(f"Input: {entry['input']}")
        print(f"Output: {entry['output']}")
        print("-" * 40)
````

# PPL

````python
import numpy as np
from openai import OpenAI

def calculate_perplexity(client, prompt, expected_output, model="google/gemma-7b-it"):
    messages = [
        {"role": "user", "content": prompt}
    ]

    response = client.chat.completions.create(
        messages=messages,
        model=model,
        max_tokens=len(expected_output.split()),  # Ensure sufficient tokens for completion
        logprobs="true",
        top_logprobs=2
    )
    print(response.choices[0])
    print(response.choices[0].message)
    token_logprobs = response.choices[0].logprobs["token_logprobs"]

    if not token_logprobs:
        raise ValueError("Log probabilities are missing from the response. Check your API support.")

    log_likelihood = sum(token_logprobs)
    perplexity = np.exp(-log_likelihood / len(token_logprobs))

    return perplexity

client = OpenAI(api_key="0", base_url="http://0.0.0.0:8000/v1")


for i in range(1,2):
    instruction = data[i]['instruction']
    input1 = data[i]['input']
    
    prompt = f"""Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.
    
    ### Instruction:
    {instruction}
    
    ### Input:
    {input1}
    
    ### Response:"""
    
    expected_output = data[i]['output']
    
    ppl = calculate_perplexity(client, prompt, expected_output)
    print(f"Perplexity: {ppl}")
````

# BLEU Score

````python
from openai import OpenAI
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import time

def calculate_bleu(client, prompt, expected_output, model="Qwen/Qwen2.5-3B-Instruct"):
    # Ensure nltk is set up for BLEU calculation

    messages = [
        {"role": "user", "content": prompt}
    ]

    response = client.chat.completions.create(
        messages=messages,
        model=model,
        max_tokens=len(expected_output.split()),  # Ensure sufficient tokens for completion
    )

    # Extract the generated output
    generated_output = response.choices[0].message.content
    # print(generated_output)
    # Tokenize the expected output and generated output
    reference = [nltk.word_tokenize(expected_output)]  # BLEU expects a list of reference sentences
    candidate = nltk.word_tokenize(generated_output)

    # Calculate BLEU score with smoothing to handle edge cases
    smooth_func = SmoothingFunction().method1
    bleu_score = sentence_bleu(reference, candidate, smoothing_function=smooth_func)

    return bleu_score

client = OpenAI(api_key="0", base_url="http://0.0.0.0:8000/v1")

result = []
start_time = time.time()
for i in range(100,200):
    instruction = data[i]['instruction']
    input1 = data[i]['input']
    
    prompt = f"""Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.
    
    ### Instruction:
    {instruction}
    
    ### Input:
    {input1}
    
    ### Response:"""
    
    expected_output = data[i]['output']

    bleu = calculate_bleu(client, prompt, expected_output)
    # print(expected_output)
    # print(f"BLEU Score: {bleu}")
    result.append(bleu)
end_time = time.time()
print("The time needed is " + str(end_time-start_time))
print(np.mean(result))
````

## Qwen2.5-3B-Instruct

````python
from tqdm import tqdm
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "Qwen/Qwen2.5-3B-Instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)

````

````python
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import time
from transformers import AutoModelForCausalLM, AutoTokenizer

def calculate_bleu(model, tokenizer, prompt, expected_output):
    # Tokenize the prompt and prepare the model inputs
    content = """Give me python code to accomplish this function. Do not include any explanation. 
                 Only provide me the code that is needed after the function definition."""
    messages = [
        {"role": "system", "content": content + "Now answer the following coding question."},
        {"role": "user", "content": prompt}
    ]

    # Create the chat input template and tokenize
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

    # Generate output
    generated_ids = model.generate(
        **model_inputs,
        max_new_tokens=len(expected_output.split())
    )
    # Decode and extract the response
    generated_output = tokenizer.batch_decode(
        generated_ids[:, model_inputs.input_ids.shape[1]:],
        skip_special_tokens=True
    )[0]

    # Tokenize reference and candidate for BLEU calculation
    reference = [nltk.word_tokenize(expected_output)]  # List of reference sentences
    candidate = nltk.word_tokenize(generated_output)

    # Calculate BLEU score
    smooth_func = SmoothingFunction().method1
    bleu_score = sentence_bleu(reference, candidate, smoothing_function=smooth_func)

    return bleu_score


result = []
start_time = time.time()
for i in range(100, 200):
    instruction = data[i]['instruction']
    input1 = data[i]['input']

    prompt = f"""Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.
    
    ### Instruction:
    {instruction}
    
    ### Input:
    {input1}
    
    ### Response:"""

    expected_output = data[i]['output']
    bleu = calculate_bleu(model, tokenizer, prompt, expected_output)
    result.append(bleu)

end_time = time.time()
print("The time needed is " + str(end_time - start_time))
print(np.mean(result))
````

````python
import re
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from transformers import AutoTokenizer, AutoModelForCausalLM

def generate_bleu_score(prompt: str, expected_output: str, max_length: int = 512) -> float:
    """
    Generates a response for a given prompt and calculates its BLEU score against the expected output.

    Args:
        prompt (str): The input prompt for code generation.
        expected_output (str): The reference output to calculate BLEU score.
        model: Pre-trained language model for text generation.
        tokenizer: Tokenizer for the pre-trained model.
        max_length (int): The maximum length of the generated sequence.

    Returns:
        float: The BLEU score of the generated response.
    """
    # Create the input prompt with instructions
    content1 = "Write the code for the following task.\n"
    content2 = "\nDo not include any explanations or examples. Provide the plain text function code only."
    messages = content1 + prompt + content2

    # Tokenize the input
    input_ids = tokenizer(messages, return_tensors="pt").to(model.device)

    # Generate the output
    outputs = model.generate(
        **input_ids,
        max_new_tokens=max_length
    )

    # Decode the generated output
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    generated_output = response.replace(messages, "").strip()

    # Extract code block if present
    code_blocks = re.findall(r"```(?:python)?(.*?)```", generated_output, re.DOTALL)
    generated_output = code_blocks[0].strip() if code_blocks else generated_output

    # Tokenize reference and candidate for BLEU calculation
    reference = [nltk.word_tokenize(expected_output)]  # List of reference sentences
    candidate = nltk.word_tokenize(generated_output)

    # Calculate BLEU score with smoothing
    smooth_func = SmoothingFunction().method1
    bleu_score = sentence_bleu(reference, candidate, smoothing_function=smooth_func)

    return bleu_score


result = []
start_time = time.time()
for i in range(100, 200):
    instruction = data[i]['instruction']
    input1 = data[i]['input']

    prompt = f"""Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.
    
    ### Instruction:
    {instruction}
    
    ### Input:
    {input1}
    
    ### Response:"""

    expected_output = data[i]['output']
    bleu = calculate_bleu(model, tokenizer, prompt, expected_output)
    result.append(bleu)

end_time = time.time()
print("The time needed is " + str(end_time - start_time))
print(np.mean(result))
````
